#### Statistical analysis

```python
from scipy.stats import kstest,anderson
```

```python
# KS Test on Total_Bill Column

h0='total_bill column follows the normal distribution'
h1='total_bill column does NOT follow normal distribution'

print(f'h0 : {h0}')
print(f'h1: {h1}\n')

x=df['total_bill']
z_val=(x - x.mean()) / x.std()
Statistic,pvalue=stats.kstest(z_val,'norm')
print(f'Statistic :{Statistic}\npvalue :{pvalue}\n')

if pvalue>0.05:
    print('Fail to reject H0 (total_bill distribution fits normal distribution)')
else:
    print('Reject H0 (total_bill distribution does NOT fit normal distribution)')
```

```python
# AD test on Total_Bill Column

h0='total_bill column follows the normal distribution'
h1='total_bill column does NOT follow normal distribution'

print(f'h0 : {h0}')
print(f'h1: {h1}\n')

x=df['total_bill']
Statistic,critical_values,significance_level =stats.anderson(x,'norm')
print(f' Statistic :{Statistic}\n critical_values:{critical_values} \n significance_level  :{significance_level }')

print('\t')

for i,j in zip(critical_values,significance_level):
    if i>Statistic:
        print(f"At {j}% significance level i.e H0 ACCEPTED (total_bill column Fits normal distribution)")
    else:
        print(f"At {j}% significance level i.e H0 REJECTED (total_bill column Not normally distributed)")
```

```python
# KS Test on tip Column

h0='tip column follows the normal distribution'
h1='tip column does NOT follow normal distribution'

print(f'h0 : {h0}')
print(f'h1: {h1}\n')

x=df['tip']
z_val=(x - x.mean()) / x.std()
Statistic,pvalue=stats.kstest(z_val,'norm')
print(f'Statistic :{Statistic}\npvalue :{pvalue}\n')

if pvalue>0.05:
    print('Fail to reject H0 (tip distribution fits normal distribution)')
else:
    print('Reject H0 (tip distribution does NOT fit normal distribution)')
```

```python
# AD test on tip Column

h0='tip column follows the normal distribution'
h1='tip column does NOT follow normal distribution'

print(f'h0 : {h0}')
print(f'h1: {h1}\n')

x=df['tip']
Statistic,critical_values,significance_level =stats.anderson(x,'norm')
print(f' Statistic :{Statistic}\n critical_values:{critical_values} \n significance_level  :{significance_level }')

print('\t')

for i,j in zip(critical_values,significance_level):
    if i>Statistic:
        print(f"At {j}% significance level i.e H0 ACCEPTED (tip column Fits normal distribution)")
    else:
        print(f"At {j}% significance level i.e H0 REJECTED (tip column Not normally distributed)")
```
